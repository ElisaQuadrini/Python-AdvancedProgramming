```python
class Node:
  def __init__(self, data): # the init method tells what to do when the assignment method is executed
    self.data = data # we are storing the input data
    self.next = None
    self.prev = None

  # additional
  def __str__(self): # the str method tells a function what to do when the string function is called = special type of string
    return f"{self.data}" # if we want to display the value of the node when we call the function print, it is automatically displaying this line of code

  # What happens if we try to apply comparison operators to nodes? We have specify what to do -> we need a criteria
  def __eq__(self, other): # this returns what to do when the equal operator is called on the node
    return self.data == other.data

node1 = Node(data = 7) # istance of the data type node
print(node1) # <__main__.Node object at 0x7db5ebe87740> : object of type node stored in a certain memory address
node2 = Node(data = 7)
print(node2)

node3 = node1

node1 == node2
```

```python
print(node1)
print(node2) # two different memory address, even if they contains the same value for data because we are looking where the node are
print(node3) # the address is the same = is like telling that what is node1 can be called even node 3

# now it makes sense this equality:
print(node1 == node3)
print(node2 == node3) # because they mean the same -> as we define in the class definition

print(node1 is node3)
```

```python
class DoublyLinkedList :
  def __init__(self,): # we start with the empty linked list so we do not need the data
    self.head = None
    self.tail = None
    self.length = 0

  def __str__(self): # in order to check if our other methods work : special string method to print the list
    s = '['
    temp = self.head
    while temp is not None:
      s = s + str(temp) + ',' # now temp is a string thanks to the definition of string in the Node definition = we have the method
      temp = temp.next # update of the pointer
    s = s + ']'
    return s

  def append(self,data): # self is the object itself
    new_node = Node(data)

    if self.length == 0:
      self.head = new_node
      self.tail = new_node

    else:
      self.tail.next = new_node
      new_node.prev = self.tail
      # self.tail.next.prev = self.tail is the second way to perform new_node.prev = self.tail
      self.tail = new_node # we need to erform this operation after attatching the new_node otherwise we would loose the reference to the tail
    self.length +=1


  def pop(self,): # removing the last element and it has to return the popped element -> nothing in input
    if self.length == 0:
      return None

    popped_element = self.tail

    # just 1 element
    if self.length == 1:
      # popped_element = self.tail wrong ! because we already saved popped_element
      self.tail = None
      self.head = None

    else:
      self.tail = popped_element.prev #self.tail.prev -> we are moving the tail to the previous element
      popped_element.prev = None
      self.tail.next = None # we break the arrow to the popped_element

    self.length -= 1

    return popped_element


dll = DoublyLinkedList()
print(dll) # empty linked list: if I didn't defined the string method, we would have the memory address

# add nodes
dll.append(45)
dll.append(76)
dll.append(5)
print(dll) # if we have done everything right, this should return the right result
```
